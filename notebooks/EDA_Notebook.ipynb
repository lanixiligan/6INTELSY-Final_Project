```python
import os
import matplotlib.pyplot as plt
from PIL import Image

# Force root directory
if os.getcwd().endswith("notebooks"): os.chdir("..")

def show_samples(category="broken_large"):
    path = os.path.join(r"data\mvtec\bottle\test", category)
    img_name = os.listdir(path)[0]
    img = Image.open(os.path.join(path, img_name))
    plt.imshow(img)
    plt.title(f"Sample: {category}")
    plt.axis('off')

plt.figure(figsize=(10,5))
plt.subplot(1,2,1); show_samples("good")
plt.subplot(1,2,2); show_samples("broken_large")
plt.show()
```
