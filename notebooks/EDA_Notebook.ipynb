```python
import os
import pandas as pd

# Iterate through all folders in the test directory to get a full count
test_path = r"data\mvtec\bottle\test"
categories = os.listdir(test_path)
counts = {cat: len(os.listdir(os.path.join(test_path, cat))) for cat in categories}

# Present as a table for the report
df_dist = pd.DataFrame(list(counts.items()), columns=['Category', 'Count'])
print(df_dist)
```

```python
import os
import matplotlib.pyplot as plt
from PIL import Image

# Force root directory
if os.getcwd().endswith("notebooks"):
    os.chdir("..")

def show_samples(category, row, num_images=5):
    path = os.path.join(r"data\mvtec\bottle\test", category)
    
    files = os.listdir(path)
    count = len(files)
    files = files[:num_images]

    for i, file in enumerate(files):
        plt.subplot(2, num_images, row * num_images + i + 1)
        img = Image.open(os.path.join(path, file))
        plt.imshow(img)
        plt.title(f"{category}\nTotal: {count}", fontsize=8)
        plt.axis('off')

plt.figure(figsize=(12,6))

show_samples("good", row=0)
show_samples("broken_large", row=1)

plt.tight_layout()
plt.show()
```
