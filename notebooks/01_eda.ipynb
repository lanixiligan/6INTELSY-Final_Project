```python
import os
import matplotlib.pyplot as plt
from PIL import Image

# 1. Force the notebook to look at the project root
if os.getcwd().endswith("notebooks"):
    os.chdir("..")

# 2. Use a 'raw' string (r"...") to fix the Windows backslash error
sample_path = r"data\mvtec\bottle\test\broken_large"

if os.path.exists(sample_path):
    # Get the first image file in the folder
    files = [f for f in os.listdir(sample_path) if f.endswith(('.png', '.jpg'))]
    if files:
        img_file = files[0]
        img = Image.open(os.path.join(sample_path, img_file))
        
        plt.figure(figsize=(8, 6))
        plt.imshow(img)
        plt.title("MVTec Bottle: Broken Large Defect")
        plt.axis('off')
        plt.show()
        print("✅ Image loaded!")
    else:
        print("❌ Folder found, but it is empty!")
else:
    print(f"❌ Still not found. Attempted path: {os.path.abspath(sample_path)}")
```
